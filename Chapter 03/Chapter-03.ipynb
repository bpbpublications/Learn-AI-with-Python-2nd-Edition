# Univariate Plots : Histograms

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data  # This contains the feature data
feature_names = iris.feature_names  # Names of the features

# Choose a specific feature (e.g., Petal Length which is the third feature)
feature_index = 2
feature_data = data[:, feature_index]

# Creating the histogram
plt.hist(feature_data, bins=15, color='lightgreen', edgecolor='black')

# Adding titles and labels
plt.title(f'Histogram of {feature_names[feature_index]}')
plt.xlabel(feature_names[feature_index])
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = iris.data  # This contains the feature data
feature_names = iris.feature_names  # Names of the features

# Convert to DataFrame for easier handling with pandas
df = pd.DataFrame(data, columns=feature_names)

# Creating Histograms for all features
plt.figure(figsize=(7.2, 5.5))  # Adjust the size for better readability
for i, feature_name in enumerate(feature_names):
    plt.subplot(2, 2, i + 1)  # Adjust subplot to fit all features
    plt.hist(df[feature_name], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature_name}')
    plt.xlabel(feature_name)
    plt.ylabel('Frequency')

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```

# Univariate Plots : Density Plots

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = iris.data  # This contains the feature data
feature_names = iris.feature_names  # Names of the features

# Convert to DataFrame for easier handling with seaborn
df = pd.DataFrame(data, columns=feature_names)

# Select the petal length feature (which is the third feature)
feature_name = feature_names[2]
petal_length_data = df[feature_name]

# Creating the density plot for petal length
plt.figure(figsize=(7.2, 3.5))
sns.kdeplot(petal_length_data, shade=True, color='purple')

# Adding titles and labels
plt.title(f'Density Plot of {feature_name}')
plt.xlabel(feature_name)
plt.ylabel('Density')

# Display the plot
plt.show()
```

# Multivariate Plots:Correlation Matrix Plot

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = iris.data  # This contains the feature data
feature_names = iris.feature_names  # Names of the features

# Convert to DataFrame for easier handling
df = pd.DataFrame(data, columns=feature_names)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Creating the correlation matrix plot
plt.figure(figsize=(7.2, 3.5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Adding title
plt.title('Correlation Matrix Plot of Iris Dataset')

# Display the plot
plt.show()
```

# Multivariate Plots:Scatter Matrix Plot

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = iris.data
feature_names = iris.feature_names
target = iris.target
target_names = iris.target_names

# Convert to DataFrame for easier handling
df = pd.DataFrame(data, columns=feature_names)
df['species'] = pd.Categorical.from_codes(target, target_names)

# Creating the scatter matrix plot
sns.pairplot(df, hue='species', diag_kind='kde', markers=['o', 's', 'D'])

# Adding title
plt.suptitle('Scatter Matrix Plot of Iris Dataset', y=1.02)  # y=1.02 to adjust title above the plot

# Display the plot
plt.show()
```

# Handling Missing Values: Imputation

```python
# Import necessary libraries
import pandas as pd
from sklearn.impute import SimpleImputer

# Sample dataset with missing values
data = {'Feature1': [1, 2, None, 4], 'Feature2': [None, 2, 3, 4]}
df = pd.DataFrame(data)

# Initialize the SimpleImputer with the mean strategy
imputer = SimpleImputer(strategy='mean')

# Apply imputation
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("Imputed Data:\n", df_imputed)
```

# Encoding Categorical Variables : One-Hot Encoding

```python
# Import necessary libraries
import pandas as pd

# Sample dataset with categorical data
df = pd.DataFrame({'Color': ['Red', 'Green', 'Blue', 'Green']})

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Color'])

print("One-Hot Encoded Data:\n", df_encoded)
```

# Encoding Categorical Variables : Label Encoding

```python
# Import necessary libraries
from sklearn.preprocessing import LabelEncoder

# Sample dataset with categorical data
df = pd.DataFrame({'Color': ['Red', 'Green', 'Blue', 'Green']})

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding
df['Color_encoded'] = label_encoder.fit_transform(df['Color'])

print("Label Encoded Data:\n", df)
```

# Scaling:Min-Max Scaling

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max scaling to the dataset
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Display the first few rows of the original and scaled data
print("Original Data:\n", df.head())
print("\nScaled Data:\n", df_scaled.head())
```

# Scaling:Standardization

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply standardization to the dataset
df_standardized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Display the first few rows of the original and standardized data
print("Original Data:\n", df.head())
print("\nStandardized Data:\n", df_standardized.head())
```

# Binarization 

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import Binarizer
import pandas as pd

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Initialize the Binarizer with a threshold value (e.g., 2.5)
binarizer = Binarizer(threshold=2.5)

# Apply binarization to the dataset
df_binarized = pd.DataFrame(binarizer.fit_transform(df), columns=df.columns)

# Display the first few rows of the original and binarized data
print("Original Data:\n", df.head())
print("\nBinarized Data:\n", df_binarized.head())
```
