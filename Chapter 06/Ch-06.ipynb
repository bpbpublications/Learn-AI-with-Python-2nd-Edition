# Building a Classifier

```python
#Step1: - Import ML library 
import sklearn
#Step2: - Import dataset
from sklearn.datasets import load_breast_cancer

data_cancer = load_breast_cancer() 


labelnames = data_cancer['target_names']
labels = data_cancer['target']
featurenames = data_cancer['feature_names']
features = data_cancer['data']

print(labelnames)


print(labels)


print(featurenames)


print(features)

# Step3: - Organizing data-training and testing set 
from sklearn.model_selection import train_test_split
train, test, train_labels, test_labels = train_test_split(features,labels,test_size = 0.30, random_state = 0)

# Step4: - Creating ML model
from sklearn.naive_bayes import GaussianNB #import the Gaussian Na√Øve Bayes model 
NB_Gaussian = GaussianNB() #initialize the model

# Step5: -Train the model 
NB_Classifier = NB_Gaussian.fit(train, train_labels)

# Step6: - Predicting test set result  
predictions = NB_Classifier.predict(test)

print(predictions)

# Step7: - Evaluating the accuracy 
from sklearn.metrics import confusion_matrix
cm_NBClassifier= confusion_matrix(test_labels, predictions)
print('Confusion Matrix:',cm_NBClassifier)


from sklearn.metrics import accuracy_score
print('Accuracy:',accuracy_score(test_labels,predictions))
```

# Building a Regressor

```python
#Step1: - Import ML library 
# Import scikit-learn
import sklearn

#Step2: -Import dataset 
# Import the dataset
from sklearn.datasets import fetch_california_housing

# Load the dataset
california_data = fetch_california_housing() 


X = california_data.data 
y = california_data.target

#Step3: -Organizing data into training and testing set 
# Organize data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Step4: -Creating ML model
# Create the model
from sklearn.linear_model import LinearRegression 
model= LinearRegression() #creating regression object 
  

#Step5: -Train the model 
model.fit(X_train, y_train) 
# Make predictions on both the training and testing data 
y_train_pred = model.predict(X_train) 
y_test_pred = model.predict(X_test)

#Step6: -Plotting Predictions against Actual Values 

import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
plt.figure(figsize=(7.2, 3.5))

# Plot predictions vs actual values for the training set
plt.scatter(y_train, y_train_pred, color='blue', label='Train Data', alpha=0.6)

# Plot predictions vs actual values for the testing set
plt.scatter(y_test, y_test_pred, color='red', label='Test Data', alpha=0.6)

# Add a diagonal line (perfect prediction line)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)

# Add labels, title, and legend
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predictions vs Actual Values (California Housing)')
plt.legend()

# Show the plot
plt.show()

#Step7: -Evaluate the model
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
print(f"Mean Squared Error (Train): {mse_train}")
print(f"Mean Squared Error (Test): {mse_test}")
```

# R Squared Example

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
Xactual = [2, -1, 8, 9]
Ypredict = [1.5, -0.7, 6, 7.9]
print ('Mean Absolute Error(MAE) =',mean_absolute_error(Xactual, Ypredict))
print ('Mean Squared Error(MSE) =',mean_squared_error(Xactual, Ypredict))
print ('Coefficient of Determination (R Squared) =',r2_score(Xactual, Ypredict))
```
